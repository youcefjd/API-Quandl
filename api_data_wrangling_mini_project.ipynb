```python
# Q3/ Calculate what the highest and lowest opening prices were for the stock in this period.
```

```python
l = []

# To answer this question I created an empty list to which I will assign (append) the second value (the column 'Open') of the dictionary d2
```

```python
for d in d2:
    q1 = d[1]
    l.append(q1)
print(max(l))

# Iterating over the dictionary d2, extracting all the values of the second column "Open" and appending them to a list
# Priting the maximum value of the list
```

```python
# Q4/ What was the largest change in any one day (based on High and Low price)?
```

```python
# Column Names: ['Date', 'Open', 'High', 'Low', 'Close', 'Change', 'Traded Volume', 'Turnover', 
#        'Last Price of the Day', 'Daily Traded Units', 'Daily Turnover']
```

```python
l1 = []
for d in d2:
    q2 = abs(d[1] - d[2])
    l1.append(q2)
print(max(l1))

# Iterating over the dictionary and appending the difference between the two columns "Open" and "Close" to a list
```

```python
# Q5/ What was the largest change between any two days (based on Closing Price)?
```

```python
l2 = []
for d in d2:
    q3 = d[4]
    l2.append(q3)
    
# Again, same concept: iterating over the dictionary and appending the relevent column to an empty list
```

```python
max([x - l2[i - 1] for i, x in enumerate(l2)][1:])

# A for loop to find the largest difference between two elements in the list "l2"
```

```python
# Q6/ What was the average daily trading volume during this year?
```

```python
l3 = []
for d in d2:
    q4 = d[6]
    l3.append(q4)
```

```python
# Average = Sum/Len. Therefore:
```

```python
s = sum(l3)
l = len(l3)
print(s/l)
```
